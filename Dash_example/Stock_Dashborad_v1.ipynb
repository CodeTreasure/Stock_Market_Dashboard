```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
```

```python
### Get Stock Data
```

```python
start_date = '2016-01-01'
end_date = '2020-05-01'
stocks = ['Amazon', 'Apple', 'Alibaba','Boeing', 'Facebook', 'Google','Microsoft', 'Netflix', 'Nvidia', 'SPY', 'Tesla']
tickers = ['AMZN', 'AAPL', 'BABA', 'BA', 'FB', 'GOOGL', 'MSFT', 'NFLX', 'NVDA' ,'SPY', 'TSLA']

data = yf.download(tickers, 
                      start=start_date, 
                      end=end_date, 
                      progress=False)
data.head()
```

```python
data = data['Adj Close']
data.tail()
```

```python
### Update the data
```

```python
def update(data):
    today = datetime.date.today()
    if data.index[-1].date() < today:
        new_data = yf.download(tickers, start=data.index[-1]+datetime.timedelta(days=1), end=today)
        new_data = new_data['Adj Close']
        return pd.concat([data, new_data])
    else: return data
```

```python
updated_data = update(data)
updated_data = round(updated_data,3)
updated_data.to_csv('stock_price.csv')
updated_data.tail()
```

```python
# Authentication to Google Drive
from googleapiclient.discovery import build
from MyAuth import *

SCOPES = "https://www.googleapis.com/auth/drive"
CLIENT_SECRET_FILE = "client_id.json"
credentials = MyAuth(client_secret_filename=CLIENT_SECRET_FILE, scopes=SCOPES).get_credentials()
drive_service = build('drive', 'v3', credentials=credentials)
```

```python
# Upload Data to Google Drive
from apiclient.http import MediaFileUpload

drive_file_name = 'stock_price.csv'
drive_file_mimeType = 'text/csv'

local_file_name = 'stock_price.csv'
local_file_mimeType = 'text/csv'

# Create the request metatdata, letting drive API know what it's receiving.
# In this example, we place the image inside the shared drive root folder,
# which has the same ID as the shared drive itself, but we could instead
# choose the ID of a folder inside the shared drive.

drive_metadata = {
        'name': drive_file_name,
        'mimeType': drive_file_mimeType,
        'parents': ['parent_file_id']}

# Now create the media file upload object and tell it what file to upload,
# in this case, "wakeupcat.jpg"

local_data = MediaFileUpload(local_file_name, mimetype=local_file_mimeType)

# Upload the file, making sure supportsAllDrives=True to enable uploading
# to shared drives.
file = drive_service.files().create(
        body=drive_metadata, media_body=local_data, supportsAllDrives=True).execute()

print("Created file '%s' id '%s'." % (file.get('name'), file.get('id')))
print(file)
```

```python
# Download Data from GoogleDrive
import io
import shutil 
from googleapiclient.http import MediaIoBaseDownload


file_id = 'file_id'
file_type = 'text/csv'
output_file = 'drive_stock_data.csv'


# download
request = drive_service.files().export(fileId=file_id, mimeType=file_type)

fh = io.BytesIO()
downloader = MediaIoBaseDownload(fh, request)
done = False
while done is False:
    status, done = downloader.next_chunk()
    print("Download %d%%" % int(status.progress() * 100))

# The file has been downloaded into RAM, now save it in a file
fh.seek(0)
with open(output_file, 'wb') as f:
    shutil.copyfileobj(fh, f, length=131072)

print("Download Stock data '%s' from Google Drive, id '%s'." % (output_file, file_id))
```

```python
### Calculate statistics
```

```python
def stats_func(data):
    stock_data = data.copy()
    tickers = stock_data.columns
    for ticker in tickers:  
#         stock_data[ticker +'_return'] = (np.log(stock_data[ticker]) - np.log(stock_data[ticker].shift(1)))*100
        stock_data[ticker +'_return'] = stock_data[ticker].pct_change()*100
        stock_data[ticker + '_momentum'] = stock_data[ticker]-stock_data[ticker].shift(12)
        stock_data[ticker +'_diff'] = stock_data[ticker].diff()
        stock_data[ticker + '_MA5'] = stock_data[ticker].rolling(5).mean()
        
    return stock_data
```

```python
### Multi-Processing
```

```python
import time
import multiprocessing
from multiprocessing import Process

updated_data = updated_data[tickers]
data_set = np.array_split(updated_data, multiprocessing.cpu_count())

p = multiprocessing.Pool(multiprocessing.cpu_count())
time_start=time.time()
res = p.map(stats_func, data_set)
time_end=time.time()
result = pd.concat(res)
p.close()
p.join()
print('time cost',round((time_end-time_start)*1000,2),'ms')
```

```python
# Corrlation Calcul
return_list = [col for col in result.columns if 'return' in col]
correlation = result[return_list].corr()

corr_df = pd.DataFrame([(i.split('_')[0],j.split('_')[0], round(correlation[i][j],2)) for i in return_list for j in return_list], columns=['stock1', 'stock2', 'corr'])
corr_df.tail()
```

```python
# Calc yearly return
yearly_return = pd.DataFrame(columns=tickers)
year_list = result.index.year.unique()

for year in year_list:
    yearly_return.loc[str(year)] = round(100*(updated_data[str(year)].iloc[-1]/updated_data[str(year)].iloc[0]-1), 2)
yearly_return.loc['total'] = round(100*(updated_data.iloc[-1]/updated_data.iloc[0]-1),2)

yearly_return.reset_index(inplace=True)
yearly_return.columns = [' ']+ stocks
yearly_return
```

```python
# Calc Portfolio
def build_hierarchical_dataframe(df, levels, value_column, color_columns=None):
    """
    Build a hierarchy of levels for Sunburst or Treemap charts.

    Levels are given starting from the bottom to the top of the hierarchy,
    ie the last level corresponds to the root.
    """
    df_all_trees = pd.DataFrame(columns=['id', 'parent', 'value', 'color'])
        
    for i, level in enumerate(levels):
        df_tree = pd.DataFrame(columns=['id', 'parent', 'value', 'color'])
        dfg = df.groupby(levels[i:]).sum()
        dfg = dfg.reset_index()
        df_tree['id'] = dfg[level].copy()
        if i < len(levels) - 1:
            df_tree['parent'] = dfg[levels[i+1]].copy()
        else:
            df_tree['parent'] = 'Portfolio'
        df_tree['value'] = dfg[value_column]
        if color_columns:
            df_tree['color'] = dfg[color_columns[0]]
            
        df_all_trees = df_all_trees.append(df_tree, ignore_index=True)
    total = pd.Series(dict(id='Portfolio', parent='',
                              value=df[value_column].sum()))
    df_all_trees = df_all_trees.append(total, ignore_index=True)
    if color_columns:
        df_all_trees['color'].fillna(df[color_columns[0]].sum(), inplace = True)
    return df_all_trees
```

```python
portf=pd.DataFrame()
portf['level_0'] = ['Equities', 'Equities','Fixed Income', 'Fixed Income', 'Commodities', 'Cash']
portf['level_1'] = ['Stocks', 'Equity Derivatives', 'Bonds','Bond Derivatives', 'Futures', '']
portf['value'] = [500, 260, 200, 140, 200, 150]
portf

# data in each level should be unique
# parents and their children cannot be the same
# levels used for the hierarchical chart
# color_columns used for the colorscale
levels = ['level_1', 'level_0'] 
value_column = 'value'
color_columns = ['value']

sunburst_df = build_hierarchical_dataframe(portf, levels, value_column, color_columns)
sunburst_df
```
